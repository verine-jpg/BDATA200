<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">B. Definition — The Curse of Dimensionality</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    The <strong>curse of dimensionality</strong> (coined by Richard Bellman, 1961) refers to a collection of phenomena that arise when data live in a space with a large number of features <em>p</em>.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Formally, consider <em>n</em> data points in &#8477;<sup>p</sup>. As <em>p</em> grows:</p>
  <ol style="color:#333; line-height:1.8;">
    <li><strong>Volume explosion:</strong> The volume of a unit hypercube grows as 1<sup>p</sup> = 1, but the volume of the inscribed hypersphere shrinks to zero. Most of the space is in the "corners".</li>
    <li><strong>Distance concentration:</strong> For random points <strong>x</strong><sub>i</sub>, the ratio (max distance − min distance) / min distance → 0 as p → ∞. All points become equidistant.</li>
    <li><strong>Sample sparsity:</strong> To maintain the same density of observations, you need exponentially more data: n ∝ k<sup>p</sup>.</li>
  </ol>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Key consequence:</strong> Algorithms that rely on local structure — nearest neighbours, kernel methods, density estimation — lose their power because "local" becomes meaningless when every point is equally far from every other point.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">C. Intuition — Why Data Become "Far Apart"</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>The Party Analogy:</strong> Imagine a small room (2D) with 20 people — it feels crowded, and you can tell who your nearest neighbours are. Now imagine the same 20 people in a giant warehouse (3D) — they spread out and it is harder to find your closest person. Now imagine a <em>hyper-warehouse</em> with 100 dimensions: the 20 people are so spread out that <em>everyone is roughly the same distance from everyone else</em>.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Practical effects:</strong>
  </p>
  <ul style="color:#333; line-height:1.8;">
    <li><strong>K-Nearest Neighbours:</strong> When all distances are equal, the "nearest" neighbour is essentially random.</li>
    <li><strong>Overfitting:</strong> With many features and few samples, a model can always find spurious patterns.</li>
    <li><strong>Irrelevant variables:</strong> Adding noisy features drowns out the signal in the useful features.</li>
    <li><strong>Computational burden:</strong> Distance calculations scale as O(np), storage scales as O(np).</li>
  </ul>
</div>

  <table style="width:100%; border-collapse: collapse; margin-top:10px;">

      <td style="padding:10px;"><strong>Best use cases for awareness</strong></td>
      <td style="padding:10px;">Any dataset where p &gt; n, genomics, NLP bag-of-words, sensor arrays, survey data with hundreds of questions</td>

    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Assumptions</strong></td>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 1.1 — Distance Concentration (Simulation)</h3>
  <p style="color:#555;"><strong>Question:</strong> How does the spread of pairwise distances change as we increase the number of dimensions from 2 to 500?</p>
</div>

````python
# Example 1.1 — Distance Concentration in High Dimensions
dims = [2, 5, 10, 50, 100, 500]
n_points = 200

fig, axes = plt.subplots(2, 3, figsize=(15, 9))
axes = axes.ravel()

for idx, d in enumerate(dims):
    X = np.random.uniform(0, 1, size=(n_points, d))
    dists = pdist(X, metric='euclidean')
    
    # Relative contrast: (max - min) / min
    contrast = (dists.max() - dists.min()) / dists.min()
    
    axes[idx].hist(dists, bins=40, color=UW_PURPLE, alpha=0.7, edgecolor='white')
    axes[idx].set_title(f'p = {d}   (contrast = {contrast:.2f})',
                        fontsize=12, fontweight='bold', color=UW_PURPLE)
    axes[idx].set_xlabel('Euclidean Distance')
    axes[idx].set_ylabel('Frequency')

fig.suptitle('Distance Concentration as Dimensions Increase',
             fontsize=16, fontweight='bold', color=UW_DARK, y=1.02)
plt.tight_layout()
plt.show()

print("\nInterpretation: As dimensions grow, the histogram narrows relative to its")
print("centre — the contrast ratio drops, meaning max and min distances converge.")
print("All points become nearly equidistant, breaking nearest-neighbour methods.")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 1.2 — High Dimensions in Practice: Caravan Insurance (86 features)</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Caravan.csv</a> — 5,822 customers described by 86 socio-demographic and product-ownership variables.</p>
  <p style="color:#555;"><strong>Question:</strong> Does KNN performance degrade when we use all 86 features vs a selected subset?</p>
</div>

````python
# Example 1.2 — KNN on Caravan: all features vs subset
caravan = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Caravan.csv')

# Binary target: 'Purchase' (Yes/No)
y = (caravan['Purchase'] == 'Yes').astype(int).values
X = caravan.drop(columns=['Purchase']).select_dtypes(include=[np.number]).values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split: first 4000 train, rest test
X_train, X_test = X_scaled[:4000], X_scaled[4000:]
y_train, y_test = y[:4000], y[4000:]

# All 85 features
knn_all = KNeighborsClassifier(n_neighbors=5)
knn_all.fit(X_train, y_train)
acc_all = knn_all.score(X_test, y_test)

# Top 10 features by variance
variances = X_scaled.var(axis=0)
top10_idx = np.argsort(variances)[-10:]
knn_sub = KNeighborsClassifier(n_neighbors=5)
knn_sub.fit(X_train[:, top10_idx], y_train)
acc_sub = knn_sub.score(X_test[:, top10_idx], y_test)

fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(['All 85 features', 'Top 10 features'],
              [acc_all, acc_sub],
              color=[UW_PURPLE, UW_GOLD], edgecolor='white', width=0.5)
ax.set_ylabel('Test Accuracy', fontsize=13)
ax.set_title('KNN Accuracy: All Features vs Top 10',
             fontsize=15, fontweight='bold', color=UW_DARK)
for b in bars:
    ax.text(b.get_x() + b.get_width()/2, b.get_height() + 0.005,
            f'{b.get_height():.3f}', ha='center', fontsize=13, fontweight='bold')
ax.set_ylim(0, 1.05)
plt.tight_layout()
plt.show()

print(f"\nWith 85 noisy features many are irrelevant, adding noise to distances.")
print(f"Reducing to 10 high-variance features can improve or maintain accuracy")
print(f"while cutting computation by ~88%.")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 1.3 — Sparsity Visualisation: College Dataset</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">College.csv</a> — 777 US colleges with 17 numeric features.</p>
  <p style="color:#555;"><strong>Question:</strong> How does adding random noise features affect the nearest-neighbour distance ratio?</p>
</div>

````python
# Example 1.3 — Adding noise features to College data
college = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/College.csv')
X_college = college.select_dtypes(include=[np.number]).values
X_college = StandardScaler().fit_transform(X_college)

added_dims = [0, 10, 50, 100, 200, 500]
nn_ratios = []

for extra in added_dims:
    if extra > 0:
        noise = np.random.normal(0, 1, size=(X_college.shape[0], extra))
        X_aug = np.hstack([X_college, noise])
    else:
        X_aug = X_college.copy()
    
    dists = pdist(X_aug)
    ratio = (dists.max() - dists.min()) / dists.min()
    nn_ratios.append(ratio)

total_dims = [X_college.shape[1] + d for d in added_dims]

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(total_dims, nn_ratios, 'o-', color=UW_PURPLE, linewidth=2.5, markersize=9)
ax.set_xlabel('Total Number of Features', fontsize=13)
ax.set_ylabel('Distance Contrast Ratio', fontsize=13)
ax.set_title('Distance Contrast Drops as Noise Features Are Added',
             fontsize=15, fontweight='bold', color=UW_DARK)
ax.axhline(1.0, color=UW_GOLD, linestyle='--', linewidth=1, label='Contrast = 1 (danger zone)')
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()

print("\nInterpretation: Starting with 17 real features, each batch of random noise")
print("dimensions lowers the contrast ratio, making it harder to distinguish")
print("near from far points — the essence of the curse.")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">B. Definitions</h2>

  <h3 style="color: #32006E;">K-Means Clustering</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    K-Means partitions <em>n</em> observations into <em>K</em> clusters by minimising the <strong>within-cluster sum of squares (WCSS)</strong>:
  </p>
  <p style="text-align:center; font-size:1.15em; color:#333;">
    WCSS = &Sigma;<sub>k=1</sub><sup>K</sup> &Sigma;<sub>x&isin;C<sub>k</sub></sub> ||x &minus; &mu;<sub>k</sub>||<sup>2</sup>
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Lloyd's Algorithm:</strong> (1) Initialise K centroids randomly. (2) Assign each point to its nearest centroid. (3) Recompute centroids as cluster means. (4) Repeat until convergence.
  </p>

  <h3 style="color: #32006E;">Hierarchical Clustering</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Agglomerative</strong> (bottom-up): Start with each point as its own cluster, then iteratively merge the two closest clusters. <strong>Divisive</strong> (top-down): Start with one cluster and recursively split. The result is a <strong>dendrogram</strong> — a tree that shows the merge/split sequence and distances.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Linkage methods:</strong> single (min distance), complete (max distance), average, Ward (minimises WCSS increase).
  </p>

  <h3 style="color: #32006E;">DBSCAN</h3>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Density-Based Spatial Clustering of Applications with Noise. A point is a <strong>core point</strong> if at least <code>min_samples</code> points lie within distance <code>eps</code>. Core points that are within eps of each other form a cluster. Points that are not core and not reachable from any core point are labelled <strong>noise</strong> (label = −1).
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">C. Intuition</h2>
  <ul style="color:#333; line-height:1.8;">
    <li><strong>K-Means</strong> is like placing K magnets on a table of iron filings — each filing snaps to the nearest magnet, and the magnets slide to the centre of their group. Repeat until stable.</li>
    <li><strong>Hierarchical</strong> is like building a family tree: at the bottom are individuals, and as you go up you group siblings, then cousins, then extended family. The dendrogram is the family tree.</li>
    <li><strong>DBSCAN</strong> is like a party: clusters form wherever people stand close together in a crowd. Loners standing far from any group are "noise".</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">D. When to Use Each Method</h2>
  <table style="width:100%; border-collapse: collapse; margin-top:10px;">
    <tr style="background-color: #4B2E83; color: white;">
      <th style="padding:10px;">Method</th>
      <th style="padding:10px;">Best For</th>
      <th style="padding:10px;">Avoid When</th>
      <th style="padding:10px;">Key Weakness</th>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>K-Means</strong></td>
      <td style="padding:10px;">Spherical, equal-size clusters; large n</td>
      <td style="padding:10px;">Irregular shapes, outliers, unknown K</td>
      <td style="padding:10px;">Must specify K; sensitive to initialisation</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Hierarchical</strong></td>
      <td style="padding:10px;">Small-to-medium n; need dendrogram; uncertain K</td>
      <td style="padding:10px;">Very large n (O(n&sup2;) or O(n&sup3;))</td>
      <td style="padding:10px;">Slow; merges are irreversible</td>
    </tr>
    <tr>
      <td style="padding:10px;"><strong>DBSCAN</strong></td>
      <td style="padding:10px;">Irregular shapes, noise/outlier detection</td>
      <td style="padding:10px;">Clusters of very different densities</td>
      <td style="padding:10px;">Sensitive to eps/min_samples; no guarantee on cluster count</td>
    </tr>
  </table>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 2.1 — K-Means Customer Segmentation: Credit Dataset</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Credit.csv</a> — 400 customers with income, credit limit, balance, etc.</p>
  <p style="color:#555;"><strong>Question:</strong> Can we segment customers into meaningful groups based on Income, Limit, and Balance?</p>
</div>

````python
# Example 2.1 — K-Means on Credit data
credit = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Credit.csv')
features = ['Income', 'Limit', 'Balance']
X = credit[features].values
X_scaled = StandardScaler().fit_transform(X)

# Elbow plot
inertias = []
K_range = range(2, 11)
for k in K_range:
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    km.fit(X_scaled)
    inertias.append(km.inertia_)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Elbow
axes[0].plot(K_range, inertias, 'o-', color=UW_PURPLE, linewidth=2.5, markersize=8)
axes[0].set_xlabel('K', fontsize=13)
axes[0].set_ylabel('Inertia (WCSS)', fontsize=13)
axes[0].set_title('Elbow Method', fontsize=14, fontweight='bold', color=UW_DARK)

# Silhouette
sil_scores = []
for k in K_range:
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = km.fit_predict(X_scaled)
    sil_scores.append(silhouette_score(X_scaled, labels))
axes[1].plot(K_range, sil_scores, 's-', color=UW_GOLD, linewidth=2.5, markersize=8)
axes[1].set_xlabel('K', fontsize=13)
axes[1].set_ylabel('Silhouette Score', fontsize=13)
axes[1].set_title('Silhouette Score', fontsize=14, fontweight='bold', color=UW_DARK)

# Final clustering with K=4
km_final = KMeans(n_clusters=4, n_init=10, random_state=42)
labels = km_final.fit_predict(X_scaled)
colors = [UW_PURPLE, UW_GOLD, '#C62828', '#388E3C']
for c in range(4):
    mask = labels == c
    axes[2].scatter(X[mask, 0], X[mask, 2], c=colors[c], label=f'Cluster {c}',
                    alpha=0.6, edgecolors='white', s=50)
axes[2].set_xlabel('Income', fontsize=13)
axes[2].set_ylabel('Balance', fontsize=13)
axes[2].set_title('K-Means Clusters (K=4)', fontsize=14, fontweight='bold', color=UW_DARK)
axes[2].legend()

plt.tight_layout()
plt.show()

print("\nInterpretation: The elbow suggests K=3 or K=4. Silhouette peaks help confirm.")
print("Clusters reveal natural customer segments: low-income/low-balance, high-income/")
print("low-balance (savers), high-income/high-balance, and moderate groups.")
````

  <ul style="color:#333; line-height:1.8;">

  <h3 style="color: #4B2E83;">Example 2.1b — Profiling &amp; Naming the Credit Clusters</h3>
  <p style="color:#555;">Continuing from Example 2.1 — let's profile each K-Means cluster and assign meaningful names.</p>

````python
# Example 2.1b — Profiling the K-Means clusters from Example 2.1
credit = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Credit.csv')
features = ['Income', 'Limit', 'Rating', 'Balance', 'Cards', 'Age']
X = credit[features].values
X_scaled = StandardScaler().fit_transform(X)

km = KMeans(n_clusters=4, n_init=10, random_state=42)
labels = km.fit_predict(X_scaled)
credit['Cluster'] = labels

# --- Step 1: Compute cluster means ---
profile = credit.groupby('Cluster')[features].mean().round(1)
print("=== Cluster Profiles (Mean Values) ===")
print(profile.to_string())
print()

# --- Step 2: Compare to global mean ---
global_mean = credit[features].mean()
profile_vs_global = (profile - global_mean).round(1)
print("=== Deviation from Global Mean ===")
print(profile_vs_global.to_string())
print()

# --- Step 3: Visualise with a heatmap of standardised means ---
profile_z = (profile - global_mean) / credit[features].std()

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Heatmap
sns.heatmap(profile_z.T, annot=True, fmt='.2f', cmap='PuOr', center=0,
            ax=axes[0], cbar_kws={'shrink': 0.8},
            xticklabels=[f'Cluster {i}' for i in range(4)])
axes[0].set_title('Cluster Profiles: Standardised Deviation from Mean',
                  fontsize=14, fontweight='bold', color=UW_DARK)
axes[0].set_ylabel('')

# Bar chart comparison
x = np.arange(len(features))
width = 0.2
cluster_colors = [UW_PURPLE, UW_GOLD, '#C62828', '#388E3C']
for c in range(4):
    axes[1].bar(x + c * width, profile_z.iloc[c], width=width,
                color=cluster_colors[c], label=f'Cluster {c}', edgecolor='white')
axes[1].set_xticks(x + 1.5 * width)
axes[1].set_xticklabels(features, rotation=30, ha='right')
axes[1].axhline(0, color='grey', linewidth=0.5, linestyle='--')
axes[1].set_ylabel('Std. Deviations from Mean', fontsize=12)
axes[1].set_title('Cluster Feature Comparison', fontsize=14,
                  fontweight='bold', color=UW_DARK)
axes[1].legend(fontsize=9)

plt.tight_layout()
plt.show()

# --- Step 4: Assign meaningful names ---
print("\n=== Suggested Cluster Names ===")
cluster_names = {}
for c in range(4):
    row = profile_z.iloc[c]
    dominant = row.abs().idxmax()
    direction = 'High' if row[dominant] > 0 else 'Low'
    print(f"\nCluster {c}: Dominant trait = {direction} {dominant}")
    print(f"  Full profile: " + ", ".join([f"{f}={'+' if row[f]>0 else ''}{row[f]:.1f}σ" for f in features]))

print("\n--- Marketing-Style Names (based on profiles above) ---")
print("  Cluster 0 → 'Steady Savers'      : moderate everything, reliable base")
print("  Cluster 1 → 'High Rollers'        : high income, high limits, big spenders")
print("  Cluster 2 → 'Credit Maximisers'   : high balance relative to income")
print("  Cluster 3 → 'Conservative Elders' : older, low balance, low utilisation")
print("\n(Your specific names may differ — the key is to examine the profile and choose")
print("a name that a marketing manager would immediately understand.)")
````

  <h3 style="color: #4B2E83;">Example 2.2 — Hierarchical Clustering: Hitters Dataset</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Hitters.csv</a> — 322 baseball players with 19 performance statistics.</p>
  <p style="color:#555;"><strong>Question:</strong> Can we group players by performance profiles using a dendrogram?</p>
</div>

````python
# Example 2.2 — Hierarchical clustering on Hitters
hitters = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Hitters.csv').dropna()
X_hit = hitters.select_dtypes(include=[np.number]).values
X_hit_scaled = StandardScaler().fit_transform(X_hit)

# Use a random subset for readability
np.random.seed(42)
idx = np.random.choice(len(X_hit_scaled), 50, replace=False)
X_sub = X_hit_scaled[idx]
names = hitters.iloc[idx].index.values

fig, axes = plt.subplots(1, 3, figsize=(20, 6))
linkage_methods = ['ward', 'complete', 'average']
link_colors = [UW_PURPLE, UW_GOLD, '#C62828']

for i, method in enumerate(linkage_methods):
    Z = linkage(X_sub, method=method)
    dendrogram(Z, ax=axes[i], leaf_rotation=90, leaf_font_size=7,
               color_threshold=0.7 * max(Z[:, 2]))
    axes[i].set_title(f'{method.title()} Linkage', fontsize=14,
                      fontweight='bold', color=UW_DARK)
    axes[i].set_xlabel('Player Index')
    axes[i].set_ylabel('Distance')

fig.suptitle('Dendrogram Comparison — Hitters Dataset',
             fontsize=16, fontweight='bold', color=UW_DARK, y=1.02)
plt.tight_layout()
plt.show()

print("\nInterpretation: Ward linkage produces the most balanced tree. Complete linkage")
print("tends to create compact clusters. Average linkage is a compromise. The height")
print("at which branches merge indicates dissimilarity between groups.")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 2.3 — DBSCAN Anomaly Detection: Default Dataset</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Default.csv</a> — 10,000 credit card holders with balance, income, and default status.</p>
  <p style="color:#555;"><strong>Question:</strong> Can DBSCAN identify unusual customers as noise points, and do they correspond to actual defaulters?</p>
</div>

````python
# Example 2.3 — DBSCAN on Default dataset
default = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Default.csv')
X_def = default[['balance', 'income']].values
X_def_scaled = StandardScaler().fit_transform(X_def)

db = DBSCAN(eps=0.3, min_samples=10)
db_labels = db.fit_predict(X_def_scaled)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# DBSCAN clusters
unique_labels = set(db_labels)
colors_db = plt.cm.Set1(np.linspace(0, 1, len(unique_labels)))
for label, col in zip(sorted(unique_labels), colors_db):
    mask = db_labels == label
    lbl = 'Noise' if label == -1 else f'Cluster {label}'
    alpha = 0.3 if label == -1 else 0.5
    axes[0].scatter(X_def[mask, 0], X_def[mask, 1], c=[col], label=lbl,
                    alpha=alpha, s=10, edgecolors='none')
axes[0].set_xlabel('Balance', fontsize=13)
axes[0].set_ylabel('Income', fontsize=13)
axes[0].set_title('DBSCAN Clusters', fontsize=14, fontweight='bold', color=UW_DARK)
axes[0].legend(fontsize=10)

# Compare noise vs actual default
is_noise = db_labels == -1
is_default = default['default'] == 'Yes'
axes[1].bar(['Noise Points', 'Non-Noise'],
            [is_default[is_noise].mean(), is_default[~is_noise].mean()],
            color=[UW_PURPLE, UW_GOLD], edgecolor='white', width=0.4)
axes[1].set_ylabel('Default Rate', fontsize=13)
axes[1].set_title('Default Rate: Noise vs Non-Noise',
                  fontsize=14, fontweight='bold', color=UW_DARK)

plt.tight_layout()
plt.show()

n_noise = (db_labels == -1).sum()
print(f"\nDBSCAN found {len(unique_labels) - (1 if -1 in unique_labels else 0)} clusters "
      f"and {n_noise} noise points ({100*n_noise/len(db_labels):.1f}%).")
print("Noise points often have unusual balance/income combinations that may")
print("correlate with higher default risk — DBSCAN as anomaly detector.")
````

  <p style="color: #E8D3FF; font-family: 'Encode Sans', 'Arial', sans-serif; font-size: 1.3em; margin: 10px 0; font-weight: 300;">Rotating Data to Reveal Maximum Variance</p>

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. Learning Objectives</h2>
  <ul style="color:#333; line-height:1.8;">
    <li>Understand PCA as finding <strong>linear combinations</strong> of the original features that maximise variance.</li>
    <li>Explain <strong>eigenvectors</strong> (directions) and <strong>eigenvalues</strong> (magnitudes) of the covariance matrix.</li>
    <li>Interpret <strong>loadings</strong> (how much each original feature contributes to a PC).</li>
    <li>Use the <strong>explained variance ratio</strong> to decide how many components to keep.</li>
    <li>Create and interpret <strong>biplots</strong>.</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">B. Definition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Given a centred data matrix <strong>X</strong> (n &times; p), PCA finds an orthogonal matrix <strong>W</strong> (p &times; p) such that:
  </p>
  <p style="text-align:center; font-size:1.15em; color:#333;">
    <strong>T</strong> = <strong>X W</strong>
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    where the columns of <strong>T</strong> (the <em>scores</em>) are uncorrelated and ordered by decreasing variance. The columns of <strong>W</strong> are the <strong>eigenvectors</strong> of the covariance matrix <strong>X<sup>T</sup>X</strong>/(n−1), and the corresponding <strong>eigenvalues</strong> &lambda;<sub>j</sub> give the variance captured by each component.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    The <strong>explained variance ratio</strong> for the j-th component is:
  </p>
  <p style="text-align:center; font-size:1.15em; color:#333;">
    EVR<sub>j</sub> = &lambda;<sub>j</sub> / &Sigma;<sub>k</sub> &lambda;<sub>k</sub>
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">C. Intuition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Imagine a cloud of points in 3D. PCA finds the <strong>best viewing angle</strong> — the direction from which the cloud looks most spread out (PC1), then the second-best direction perpendicular to the first (PC2), and so on. It's like rotating a 3D object on your desk until you see its shape most clearly.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Loadings</strong> tell you which original features contribute to each direction. A biplot shows both the data points (scores) and the feature arrows (loadings) in the same 2D view.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">D. When to Use PCA</h2>
  <table style="width:100%; border-collapse: collapse; margin-top:10px;">
    <tr style="background-color: #4B2E83; color: white;">
      <th style="padding:10px; text-align:left;">Aspect</th>
      <th style="padding:10px; text-align:left;">Details</th>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Best use cases</strong></td>
      <td style="padding:10px;">Feature reduction, visualisation of high-d data, pre-processing before clustering/regression, noise reduction</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Assumptions</strong></td>
      <td style="padding:10px;">Linear relationships, features should be scaled, large variance = important</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Strengths</strong></td>
      <td style="padding:10px;">Fast (eigendecomposition), deterministic, interpretable loadings</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Limitations</strong></td>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 3.1 — PCA Scree Plot &amp; Biplot: College Dataset</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">College.csv</a> — 777 US colleges with 17 numeric features.</p>
  <p style="color:#555;"><strong>Question:</strong> How many principal components are needed to capture most of the variance? Which features drive each component?</p>
</div>

````python
# Example 3.1 — PCA on College data
college = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/College.csv')
X_col = college.select_dtypes(include=[np.number])
feature_names = X_col.columns.tolist()
X_col_scaled = StandardScaler().fit_transform(X_col)

pca = PCA()
scores = pca.fit_transform(X_col_scaled)

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Scree plot
axes[0].bar(range(1, len(pca.explained_variance_ratio_)+1),
            pca.explained_variance_ratio_, color=UW_PURPLE, alpha=0.7, edgecolor='white')
axes[0].plot(range(1, len(pca.explained_variance_ratio_)+1),
             np.cumsum(pca.explained_variance_ratio_), 'o-', color=UW_GOLD, linewidth=2)
axes[0].axhline(0.9, color='#C62828', linestyle='--', linewidth=1, label='90% threshold')
axes[0].set_xlabel('Principal Component', fontsize=12)
axes[0].set_ylabel('Variance Explained', fontsize=12)
axes[0].set_title('Scree Plot', fontsize=14, fontweight='bold', color=UW_DARK)
axes[0].legend(fontsize=10)

# Biplot
is_private = (college['Private'] == 'Yes').values
for val, color, label in [(True, UW_PURPLE, 'Private'), (False, UW_GOLD, 'Public')]:
    mask = is_private == val
    axes[1].scatter(scores[mask, 0], scores[mask, 1], c=color, label=label,
                    alpha=0.5, s=30, edgecolors='white')

# Loading arrows
loadings = pca.components_[:2].T
scale = 4
for i, feat in enumerate(feature_names):
    axes[1].annotate('', xy=(loadings[i, 0]*scale, loadings[i, 1]*scale),
                     xytext=(0, 0),
                     arrowprops=dict(arrowstyle='->', color='#C62828', lw=1.5))
    axes[1].text(loadings[i, 0]*scale*1.1, loadings[i, 1]*scale*1.1,
                 feat, fontsize=7, color='#C62828')
axes[1].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%})', fontsize=12)
axes[1].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%})', fontsize=12)
axes[1].set_title('Biplot', fontsize=14, fontweight='bold', color=UW_DARK)
axes[1].legend(fontsize=10)

# Cumulative variance
n_for_90 = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= 0.9) + 1
axes[2].fill_between(range(1, len(pca.explained_variance_ratio_)+1),
                     np.cumsum(pca.explained_variance_ratio_),
                     alpha=0.3, color=UW_PURPLE)
axes[2].plot(range(1, len(pca.explained_variance_ratio_)+1),
             np.cumsum(pca.explained_variance_ratio_), 'o-', color=UW_PURPLE, linewidth=2)
axes[2].axvline(n_for_90, color=UW_GOLD, linestyle='--', linewidth=2,
                label=f'{n_for_90} PCs for 90%')
axes[2].set_xlabel('Number of Components', fontsize=12)
axes[2].set_ylabel('Cumulative Variance', fontsize=12)
axes[2].set_title('Cumulative Explained Variance', fontsize=14, fontweight='bold', color=UW_DARK)
axes[2].legend(fontsize=10)

plt.tight_layout()
plt.show()

print(f"\n{n_for_90} components capture 90% of the variance (down from {len(feature_names)} features).")
print("The biplot shows Private vs Public colleges separate along PC1, driven by")
print("enrolment size, acceptance rates, and expenditure features.")
````

  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Just as clusters must be profiled and named, <strong>principal components must be interpreted</strong>. A component called "PC1" is meaningless — but a component you can describe as <em>"Overall institutional size"</em> or <em>"Academic selectivity"</em> tells a rich story.
  </p>
  <h3 style="color: #32006E;">How to Interpret a Principal Component</h3>
  <ol style="color:#333; line-height:1.8;">
    <li><strong>Examine the loadings:</strong> Which original features have the largest positive and negative weights on this PC?</li>
    <li><strong>Look for thematic coherence:</strong> Do the high-loading features share a common concept?</li>
    <li><strong>Name the component</strong> after the concept it represents — not after the variables.</li>
    <li><strong>Check both ends:</strong> A positive score on "Academic quality" means high quality; a negative score means low quality. Both ends must make sense.</li>
  </ol>

````python
# Example 3.1b — Interpreting and naming PCA components
college = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/College.csv')
X_col = college.select_dtypes(include=[np.number])
feature_names = X_col.columns.tolist()
X_col_scaled = StandardScaler().fit_transform(X_col)

pca = PCA()
scores = pca.fit_transform(X_col_scaled)

# Build a loadings table for the first 4 PCs
loadings_df = pd.DataFrame(
    pca.components_[:4].T,
    index=feature_names,
    columns=['PC1', 'PC2', 'PC3', 'PC4']
)

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

for idx, (pc, ax) in enumerate(zip(['PC1', 'PC2', 'PC3', 'PC4'], axes.ravel())):
    vals = loadings_df[pc].sort_values()
    colors = [UW_PURPLE if v > 0 else UW_GOLD for v in vals]
    ax.barh(vals.index, vals.values, color=colors, edgecolor='white')
    ax.set_title(f'{pc} Loadings ({pca.explained_variance_ratio_[idx]:.1%} var)',
                 fontsize=13, fontweight='bold', color=UW_DARK)
    ax.axvline(0, color='grey', linewidth=0.5)
    ax.set_xlabel('Loading Weight')

plt.tight_layout()
plt.show()

# Interpretation and naming
print("=== Component Interpretation ===\n")
print("PC1: Loads heavily on Apps, Accept, Enroll, F.Undergrad →")
print("     Name: \"Institutional Size\"")
print("     High scores = large universities; Low scores = small colleges\n")

print("PC2: Loads on Top10perc, Top25perc, PhD, Terminal, Outstate →")
print("     Name: \"Academic Prestige / Selectivity\"")
print("     High scores = elite colleges; Low scores = open-access\n")

print("PC3: Loads on Personal, S.F.Ratio, Books →")
print("     Name: \"Student Self-Funding\"")
print("     High scores = students pay more out of pocket\n")

print("PC4: Loads on Room.Board, Expend →")
print("     Name: \"Cost of Attendance\"")
print("     High scores = expensive campus experience\n")

print("These names transform abstract PCs into concepts a university")
print("administrator would immediately understand and act upon.")
````

  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Boston.csv</a> — 506 census tracts with 13 features predicting median home value.</p>
  <p style="color:#555;"><strong>Question:</strong> How does reducing to 5 PCs affect cluster quality compared to using all 13 features?</p>

````python
# Example 3.2 — PCA then K-Means on Boston
boston = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Boston.csv')
X_bos = boston.drop(columns=['medv'], errors='ignore').select_dtypes(include=[np.number]).values
X_bos_scaled = StandardScaler().fit_transform(X_bos)

pca5 = PCA(n_components=5)
X_pca5 = pca5.fit_transform(X_bos_scaled)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for ax, X_in, title in [(axes[0], X_bos_scaled, f'All {X_bos_scaled.shape[1]} Features'),
                          (axes[1], X_pca5, '5 Principal Components')]:
    km = KMeans(n_clusters=3, n_init=10, random_state=42)
    labels = km.fit_predict(X_in)
    sil = silhouette_score(X_in, labels)
    
    pca2 = PCA(n_components=2)
    X_2d = pca2.fit_transform(X_in)
    
    colors = [UW_PURPLE, UW_GOLD, '#C62828']
    for c in range(3):
        mask = labels == c
        ax.scatter(X_2d[mask, 0], X_2d[mask, 1], c=colors[c], label=f'Cluster {c}',
                   alpha=0.6, s=30, edgecolors='white')
    ax.set_title(f'{title}\nSilhouette = {sil:.3f}', fontsize=13,
                 fontweight='bold', color=UW_DARK)
    ax.set_xlabel('Dim 1')
    ax.set_ylabel('Dim 2')
    ax.legend(fontsize=9)

plt.tight_layout()
plt.show()

print(f"\nPCA reduced the feature space from {X_bos_scaled.shape[1]} to 5 dimensions")
print(f"while capturing {pca5.explained_variance_ratio_.sum():.1%} of the variance.")
print("Cluster quality (silhouette) is often comparable or better after PCA,")
print("because PCA removes noise dimensions.")
````

  <ul style="color:#333; line-height:1.8;">
    <li>Understand MDS as an optimisation that places points in low-d space while preserving pairwise distances.</li>
    <li>Distinguish <strong>metric MDS</strong> (preserve exact distances) from <strong>non-metric MDS</strong> (preserve rank order).</li>
    <li>Interpret the <strong>stress</strong> value as a measure of distortion.</li>
    <li>Know when MDS is preferable to PCA (e.g., when you only have a distance matrix, not raw features).</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">B. Definition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Given an n &times; n distance matrix <strong>D</strong> with entries d<sub>ij</sub>, MDS finds coordinates <strong>y</strong><sub>1</sub>, ..., <strong>y</strong><sub>n</sub> in &#8477;<sup>k</sup> (usually k = 2) that minimise the <strong>stress</strong> function:
  </p>
  <p style="text-align:center; font-size:1.15em; color:#333;">
    Stress = &radic;[ &Sigma;<sub>i&lt;j</sub> (d<sub>ij</sub> &minus; ||y<sub>i</sub> &minus; y<sub>j</sub>||)<sup>2</sup> / &Sigma;<sub>i&lt;j</sub> d<sub>ij</sub><sup>2</sup> ]
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Metric MDS</strong> uses the actual distances. <strong>Non-metric MDS</strong> only requires that the rank order of distances is preserved (useful for ordinal or subjective similarity data).
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">C. Intuition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Imagine you have a table of driving distances between 10 cities, but no map. MDS takes that distance table and reconstructs a 2D map where cities that are far apart in driving distance are far apart on the map, and cities that are close are plotted close together. The "stress" tells you how much distortion the flat map introduces.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">D. When to Use MDS</h2>
  <table style="width:100%; border-collapse: collapse; margin-top:10px;">
    <tr style="background-color: #4B2E83; color: white;">
      <th style="padding:10px; text-align:left;">Aspect</th>
      <th style="padding:10px; text-align:left;">Details</th>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Best use cases</strong></td>
      <td style="padding:10px;">You have only a distance/dissimilarity matrix (no raw features); perceptual mapping; product similarity</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Assumptions</strong></td>
      <td style="padding:10px;">Metric MDS assumes distances are meaningful; non-metric only requires rank order</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Strengths</strong></td>
      <td style="padding:10px;">Works directly from distances; intuitive output; stress gives quality measure</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Limitations</strong></td>
      <td style="padding:10px;">O(n&sup2;) memory for distance matrix; stress can have local minima; no loadings for interpretation</td>
    </tr>
    <tr>
      <td style="padding:10px;"><strong>Common mistakes</strong></td>
      <td style="padding:10px;">Not checking stress value; using on very large n without subsampling; confusing with PCA</td>
    </tr>
  </table>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 4.1 — MDS on Credit Customer Similarity</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Credit.csv</a></p>
  <p style="color:#555;"><strong>Question:</strong> Can MDS reveal customer groups when we start from a distance matrix rather than raw features?</p>
</div>

````python
# Example 4.1 — MDS on Credit dataset
credit = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Credit.csv')
X_cr = credit[['Income', 'Limit', 'Balance', 'Rating']].values
X_cr_scaled = StandardScaler().fit_transform(X_cr)

# Compute pairwise distance matrix
D = squareform(pdist(X_cr_scaled))

# Metric MDS
mds_metric = MDS(n_components=2, dissimilarity='precomputed', random_state=42, normalized_stress='auto')
Y_metric = mds_metric.fit_transform(D)

# Non-metric MDS
mds_nonmetric = MDS(n_components=2, metric=False, dissimilarity='precomputed',
                     random_state=42, normalized_stress='auto')
Y_nonmetric = mds_nonmetric.fit_transform(D)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for ax, Y, title, stress in [(axes[0], Y_metric, 'Metric MDS', mds_metric.stress_),
                               (axes[1], Y_nonmetric, 'Non-Metric MDS', mds_nonmetric.stress_)]:
    sc = ax.scatter(Y[:, 0], Y[:, 1], c=credit['Balance'], cmap='PuOr',
                    alpha=0.6, s=30, edgecolors='white')
    ax.set_title(f'{title}\nStress = {stress:.2f}', fontsize=14,
                 fontweight='bold', color=UW_DARK)
    ax.set_xlabel('MDS Dim 1')
    ax.set_ylabel('MDS Dim 2')
    plt.colorbar(sc, ax=ax, label='Balance')

plt.tight_layout()
plt.show()

print("\nInterpretation: MDS maps the 4D credit data into 2D while preserving distances.")
print("The colour gradient (Balance) shows a clear structure — high-balance customers")
print("cluster separately from low-balance ones, confirming the distance matrix captures")
print("meaningful customer similarity.")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 4.2 — MDS vs PCA Comparison: Auto Dataset</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Auto.csv</a></p>
  <p style="color:#555;"><strong>Question:</strong> Do MDS and PCA produce similar 2D maps for the Auto dataset?</p>
</div>

````python
# Example 4.2 — MDS vs PCA on Auto
auto = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Auto.csv', na_values='?').dropna()
cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']
X_auto = auto[cols].values.astype(float)
X_auto_scaled = StandardScaler().fit_transform(X_auto)

# PCA
pca_auto = PCA(n_components=2)
Y_pca = pca_auto.fit_transform(X_auto_scaled)

# Metric MDS
D_auto = squareform(pdist(X_auto_scaled))
mds_auto = MDS(n_components=2, dissimilarity='precomputed', random_state=42, normalized_stress='auto')
Y_mds = mds_auto.fit_transform(D_auto)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
origin = auto['origin'].values
cmap = {1: UW_PURPLE, 2: UW_GOLD, 3: '#C62828'}
labels_map = {1: 'American', 2: 'European', 3: 'Japanese'}

for ax, Y, title in [(axes[0], Y_pca, 'PCA'), (axes[1], Y_mds, 'Metric MDS')]:
    for o in [1, 2, 3]:
        mask = origin == o
        ax.scatter(Y[mask, 0], Y[mask, 1], c=cmap[o], label=labels_map[o],
                   alpha=0.6, s=40, edgecolors='white')
    ax.set_title(title, fontsize=14, fontweight='bold', color=UW_DARK)
    ax.set_xlabel('Dim 1')
    ax.set_ylabel('Dim 2')
    ax.legend(fontsize=10)

plt.tight_layout()
plt.show()

print("\nInterpretation: With Euclidean distances, metric MDS and PCA produce very")
print("similar layouts (up to rotation/reflection). Both separate American (large,")
print("low-mpg) from Japanese (small, high-mpg) cars. MDS is more flexible because")
print("it works from any distance matrix, not just Euclidean.")
````

````python
# Example 4.3 — MDS on OJ dataset
oj = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/OJ.csv')
X_oj = oj.select_dtypes(include=[np.number]).values
X_oj_scaled = StandardScaler().fit_transform(X_oj)

# Subsample for speed
np.random.seed(42)
idx = np.random.choice(len(X_oj_scaled), 300, replace=False)
X_oj_sub = X_oj_scaled[idx]
purchase = oj['Purchase'].values[idx]

D_oj = squareform(pdist(X_oj_sub))
mds_oj = MDS(n_components=2, dissimilarity='precomputed', random_state=42, normalized_stress='auto')
Y_oj = mds_oj.fit_transform(D_oj)

fig, ax = plt.subplots(figsize=(10, 7))
for brand, color in [('CH', UW_PURPLE), ('MM', UW_GOLD)]:
    mask = purchase == brand
    ax.scatter(Y_oj[mask, 0], Y_oj[mask, 1], c=color, label=brand,
               alpha=0.6, s=50, edgecolors='white')
ax.set_xlabel('MDS Dim 1', fontsize=13)
ax.set_ylabel('MDS Dim 2', fontsize=13)
ax.set_title(f'MDS Map of OJ Purchases\nStress = {mds_oj.stress_:.2f}',
             fontsize=15, fontweight='bold', color=UW_DARK)
ax.legend(fontsize=12, title='Brand')
plt.tight_layout()
plt.show()

print("\nInterpretation: MDS reveals partial separation between Citrus Hill (CH) and")
print("Minute Maid (MM) purchasers, driven by pricing and promotion differences.")
print("Overlap indicates that some purchase occasions look similar regardless of brand.")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. Learning Objectives</h2>
  <ul style="color:#333; line-height:1.8;">
    <li>Understand correspondence analysis as PCA for categorical data (contingency tables).</li>
    <li>Explain <strong>row profiles</strong>, <strong>column profiles</strong>, and <strong>chi-square distances</strong>.</li>
    <li>Interpret a <strong>symmetric map</strong> (biplot) showing both row and column categories.</li>
    <li>Know when CA is appropriate and when to use alternatives.</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">B. Definition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Given a contingency table <strong>N</strong> (r rows &times; c columns), CA decomposes the <strong>standardised residuals</strong> from independence using SVD:
  </p>
  <ol style="color:#333; line-height:1.8;">
    <li>Compute the <strong>correspondence matrix</strong>: P = N / n (where n = total count).</li>
    <li>Compute row masses <strong>r</strong> = row sums of P, column masses <strong>c</strong> = column sums of P.</li>
    <li>Compute standardised residuals: S<sub>ij</sub> = (P<sub>ij</sub> &minus; r<sub>i</sub>c<sub>j</sub>) / &radic;(r<sub>i</sub>c<sub>j</sub>).</li>
    <li>Apply SVD to <strong>S</strong> to get row and column coordinates in a low-dimensional space.</li>
  </ol>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    The <strong>total inertia</strong> equals Pearson's &chi;&sup2;/n and measures the strength of the row-column association.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">C. Intuition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Think of a survey where respondents (rows) rated products (columns). CA produces a 2D map where:
  </p>
  <ul style="color:#333; line-height:1.8;">
    <li>Respondents who gave <em>similar ratings</em> appear close together.</li>
    <li>Products that received <em>similar ratings</em> appear close together.</li>
    <li>A respondent near a product means that respondent rated that product higher than average.</li>
  </ul>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    It's like PCA, but for <em>cross-tabulated categorical</em> data.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">D. When to Use CA</h2>
  <table style="width:100%; border-collapse: collapse; margin-top:10px;">
    <tr style="background-color: #4B2E83; color: white;">
      <th style="padding:10px; text-align:left;">Aspect</th>
      <th style="padding:10px; text-align:left;">Details</th>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Best use cases</strong></td>
      <td style="padding:10px;">Cross-tabulations, survey responses, demographics &times; preferences, text analysis (word &times; document)</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Assumptions</strong></td>
      <td style="padding:10px;">Data are counts in a contingency table; non-negative; chi-square distance is appropriate</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Strengths</strong></td>
      <td style="padding:10px;">Handles categorical data natively; produces interpretable joint maps; scale-free</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Limitations</strong></td>
      <td style="padding:10px;">Only for two-way tables (use MCA for more); sensitive to rare categories; distances between row and column points require careful interpretation</td>
    </tr>
    <tr>
      <td style="padding:10px;"><strong>Common mistakes</strong></td>
      <td style="padding:10px;">Interpreting row-to-column distances literally (only row-row and col-col distances are valid); ignoring low-inertia axes</td>
    </tr>
  </table>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 5.1 — CA on Wage Education &times; Job Class</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Wage.csv</a> — 3,000 workers with education level, job class, and wage.</p>
  <p style="color:#555;"><strong>Question:</strong> How are education levels associated with job classes (Industrial vs Information)?</p>
</div>

````python
# Example 5.1 — Correspondence Analysis on Wage
wage = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Wage.csv')

# Build contingency table: education × marital status (both have 5 levels)
ct = pd.crosstab(wage['education'], wage['maritl'])
print("Contingency Table:")
print(ct)
print()

# Run CA using prince
ca = prince.CA(n_components=2, random_state=42)
ca = ca.fit(ct)

fig, ax = plt.subplots(figsize=(10, 8))

# Row coordinates (education levels)
row_coords = ca.row_coordinates(ct)
ax.scatter(row_coords.iloc[:, 0], row_coords.iloc[:, 1],
           c=UW_PURPLE, s=150, zorder=5, edgecolors='white')
for idx_name in row_coords.index:
    ax.annotate(str(idx_name),
                (row_coords.loc[idx_name].iloc[0], row_coords.loc[idx_name].iloc[1]),
                fontsize=10, fontweight='bold', color=UW_PURPLE, ha='center', va='bottom',
                xytext=(0, 10), textcoords='offset points')

# Column coordinates (marital status)
col_coords = ca.column_coordinates(ct)
ax.scatter(col_coords.iloc[:, 0], col_coords.iloc[:, 1],
           c=UW_GOLD, s=150, marker='s', zorder=5, edgecolors='white')
for idx_name in col_coords.index:
    ax.annotate(str(idx_name),
                (col_coords.loc[idx_name].iloc[0], col_coords.loc[idx_name].iloc[1]),
                fontsize=10, fontweight='bold', color='#85714D', ha='center', va='bottom',
                xytext=(0, 10), textcoords='offset points')

ax.axhline(0, color='grey', linewidth=0.5, linestyle='--')
ax.axvline(0, color='grey', linewidth=0.5, linestyle='--')
ax.set_xlabel(f'Dimension 1 ({ca.percentage_of_variance_[0]:.1f}%)', fontsize=13)
ax.set_ylabel(f'Dimension 2 ({ca.percentage_of_variance_[1]:.1f}%)', fontsize=13)
ax.set_title('Correspondence Analysis: Education × Marital Status',
             fontsize=15, fontweight='bold', color=UW_DARK)
ax.legend(['Education Level', 'Marital Status'], fontsize=11)

plt.tight_layout()
plt.show()

print("\nInterpretation: Education levels close to a marital status on the map are")
print("disproportionately associated. The map reveals demographic patterns — for")
print("example, advanced degree holders may cluster near certain marital categories,")
print("reflecting age and socioeconomic correlations in the data.")
````

  <h3 style="color: #4B2E83;">Example 5.2 — CA on Heart Disease: Chest Pain × Thalassemia</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Heart.csv</a></p>

  <p style="color: #E8D3FF; font-family: 'Encode Sans', 'Arial', sans-serif; font-size: 1.3em; margin: 10px 0; font-weight: 300;">Keeping Neighbours Close, Allowing Global Distortion</p>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. Learning Objectives</h2>
  <ul style="color:#333; line-height:1.8;">
    <li>Understand t-SNE as a <strong>nonlinear</strong> dimensionality reduction technique focused on <strong>local neighbourhood preservation</strong>.</li>
    <li>Explain how t-SNE converts pairwise similarities to probabilities and minimises <strong>KL divergence</strong>.</li>
    <li>Understand the role of <strong>perplexity</strong> and its effect on the output.</li>
    <li>Know the key warnings: <strong>randomness</strong>, <strong>artificial clusters</strong>, and <strong>unreliable inter-cluster distances</strong>.</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">B. Definition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    t-SNE (van der Maaten & Hinton, 2008) works in two steps:
  </p>
  <ol style="color:#333; line-height:1.8;">
    <li><strong>High-dimensional similarities:</strong> For each pair (i, j), compute a Gaussian-based conditional probability p<sub>j|i</sub> that i would pick j as its neighbour. The <strong>perplexity</strong> parameter controls the effective number of neighbours (bandwidth of the Gaussian).</li>
    <li><strong>Low-dimensional similarities:</strong> Place points in 2D and compute similarities using a <strong>Student's t-distribution</strong> (1 degree of freedom) — heavier tails than a Gaussian, which helps prevent crowding.</li>
    <li><strong>Optimisation:</strong> Minimise the <strong>Kullback-Leibler divergence</strong> between the high-d and low-d similarity distributions using gradient descent.</li>
  </ol>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    The t-distribution in the low-d space means that moderately distant points in high-d can be placed <em>far apart</em> in 2D without paying a large cost — this creates visually clear clusters.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">C. Intuition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Imagine a crowded dance floor (high-dimensional space). t-SNE takes a photo from above, then rearranges people in a 2D seating chart so that <strong>people who were dancing close together remain seated close together</strong>. However, the distances between <em>different groups</em> of dancers in the seating chart are meaningless — they're just arranged for visual clarity.
  </p>
  <div style="background-color: #FFF3E0; border: 2px solid #F57C00; padding: 15px; border-radius: 5px; margin-top:15px;">
    <p style="margin:0; color:#C62828; font-weight:bold;">Important Warnings:</p>
    <ul style="color:#333; line-height:1.8; margin-top:5px;">
      <li>Clusters in t-SNE plots may be <strong>artificial</strong> — artefacts of the algorithm, not real structure.</li>
      <li><strong>Distances between clusters are unreliable</strong> — two clusters far apart in the plot may not be far apart in reality.</li>
      <li>Results depend on the <strong>random seed</strong> — always run multiple times.</li>
      <li><strong>Perplexity</strong> must be tuned — too low creates many tiny spurious clusters; too high blurs real structure.</li>
    </ul>
  </div>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">D. When to Use t-SNE</h2>
  <table style="width:100%; border-collapse: collapse; margin-top:10px;">
    <tr style="background-color: #4B2E83; color: white;">
      <th style="padding:10px; text-align:left;">Aspect</th>
      <th style="padding:10px; text-align:left;">Details</th>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Best use cases</strong></td>
      <td style="padding:10px;">Exploratory visualisation of high-d data; checking if clusters exist; biological/genomic data</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Assumptions</strong></td>
      <td style="padding:10px;">Local structure matters more than global; data have meaningful neighbourhoods</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Strengths</strong></td>
      <td style="padding:10px;">Reveals nonlinear structure; beautiful visualisations; works well for moderate n</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Limitations</strong></td>
      <td style="padding:10px;">Slow (O(n&sup2;)); non-deterministic; no inverse transform; global structure lost</td>
    </tr>
    <tr>
      <td style="padding:10px;"><strong>Common mistakes</strong></td>
      <td style="padding:10px;">Reading too much into cluster sizes/distances; using a single random seed; not trying multiple perplexities</td>
    </tr>
  </table>
</div>

  <h3 style="color: #4B2E83;">Example 6.1 — t-SNE Perplexity Comparison: College Dataset</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">College.csv</a></p>
  <p style="color:#555;"><strong>Question:</strong> How does perplexity affect the t-SNE visualisation of US colleges?</p>
</div>

````python
# Example 6.1 — t-SNE with different perplexities on College
college = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/College.csv')
X_col = StandardScaler().fit_transform(college.select_dtypes(include=[np.number]))
is_private = (college['Private'] == 'Yes').values

perplexities = [5, 30, 50, 100]
fig, axes = plt.subplots(1, 4, figsize=(22, 5))

for ax, perp in zip(axes, perplexities):
    tsne = TSNE(n_components=2, perplexity=perp, random_state=42, max_iter=1000)
    Y = tsne.fit_transform(X_col)
    
    for val, color, label in [(True, UW_PURPLE, 'Private'), (False, UW_GOLD, 'Public')]:
        mask = is_private == val
        ax.scatter(Y[mask, 0], Y[mask, 1], c=color, label=label, alpha=0.5, s=20)
    ax.set_title(f'Perplexity = {perp}', fontsize=13, fontweight='bold', color=UW_DARK)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.legend(fontsize=8)

fig.suptitle('t-SNE: Effect of Perplexity on College Data',
             fontsize=16, fontweight='bold', color=UW_DARK, y=1.02)
plt.tight_layout()
plt.show()

print("\nInterpretation: Low perplexity (5) creates many small, fragmented clusters —")
print("this is an artefact, not real structure. Higher perplexity (30–50) reveals the")
print("Private/Public separation more clearly. Very high perplexity (100) begins to")
print("blur the structure. Perplexity ≈ 30 is usually a good starting point.")
````

````python
# Example 6.2 — t-SNE vs PCA on Heart
heart = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Heart.csv').dropna()
X_h = StandardScaler().fit_transform(heart.select_dtypes(include=[np.number]))
y_h = (heart['AHD'] == 'Yes').values

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# PCA
pca_h = PCA(n_components=2)
Y_pca_h = pca_h.fit_transform(X_h)

# t-SNE
tsne_h = TSNE(n_components=2, perplexity=30, random_state=42)
Y_tsne_h = tsne_h.fit_transform(X_h)

for ax, Y, title in [(axes[0], Y_pca_h, 'PCA'), (axes[1], Y_tsne_h, 't-SNE (perp=30)')]:
    for val, color, label in [(True, '#C62828', 'Disease'), (False, '#388E3C', 'No Disease')]:
        mask = y_h == val
        ax.scatter(Y[mask, 0], Y[mask, 1], c=color, label=label, alpha=0.5, s=30)
    ax.set_title(title, fontsize=14, fontweight='bold', color=UW_DARK)
    ax.legend(fontsize=10)
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()

print("\nInterpretation: PCA shows overlap between disease/no-disease groups because")
print("the separation is nonlinear. t-SNE reveals clearer local clusters within each")
print("group, though some overlap remains. Remember: t-SNE cluster separation may be")
print("exaggerated — always validate with quantitative methods.")
````

  <h3 style="color: #4B2E83;">Example 6.3 — t-SNE Instability: Random Seeds on Bikeshare</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Bikeshare.csv</a></p>
  <p style="color:#555;"><strong>Question:</strong> How much does the t-SNE output change with different random seeds?</p>
</div>

````python
# Example 6.3 — t-SNE instability with different random seeds
bike = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Bikeshare.csv')
X_bike = StandardScaler().fit_transform(bike.select_dtypes(include=[np.number]))

# Subsample for speed
np.random.seed(42)
idx = np.random.choice(len(X_bike), 500, replace=False)
X_bike_sub = X_bike[idx]
season = bike['season'].values[idx]

seeds = [0, 42, 123, 999]
fig, axes = plt.subplots(1, 4, figsize=(22, 5))

season_colors = {1: UW_PURPLE, 2: UW_GOLD, 3: '#388E3C', 4: '#C62828'}
season_names = {1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'}

for ax, seed in zip(axes, seeds):
    tsne = TSNE(n_components=2, perplexity=30, random_state=seed)
    Y = tsne.fit_transform(X_bike_sub)
    
    for s in [1, 2, 3, 4]:
        mask = season == s
        ax.scatter(Y[mask, 0], Y[mask, 1], c=season_colors[s],
                   label=season_names[s], alpha=0.5, s=20)
    ax.set_title(f'seed = {seed}', fontsize=13, fontweight='bold', color=UW_DARK)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.legend(fontsize=8)

fig.suptitle('t-SNE: Different Random Seeds on Bikeshare Data',
             fontsize=16, fontweight='bold', color=UW_DARK, y=1.02)
plt.tight_layout()
plt.show()

print("\nInterpretation: Each seed produces a different arrangement! The local clusters")
print("are roughly consistent, but their positions change. This shows why you should")
print("NEVER rely on a single t-SNE run — always try multiple seeds and look for")
print("patterns that are consistent across runs.")
````

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">B. Definition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    UMAP (McInnes, Healy & Melville, 2018) is grounded in <strong>Riemannian geometry</strong> and <strong>algebraic topology</strong>, but the practical intuition is:
  </p>
  <ol style="color:#333; line-height:1.8;">
    <li><strong>Build a weighted graph</strong> in high-d: Connect each point to its <code>n_neighbors</code> nearest neighbours. Edge weights decay with distance.</li>
    <li><strong>Optimise a low-d layout</strong>: Find 2D coordinates that minimise the <strong>cross-entropy</strong> between the high-d and low-d edge weight distributions.</li>
  </ol>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Key parameters:</p>
  <ul style="color:#333; line-height:1.8;">
    <li><code>n_neighbors</code>: Controls local vs global balance. Small = focus on local structure. Large = more global context.</li>
    <li><code>min_dist</code>: Controls how tightly points can pack. Small = tight clusters. Large = more spread out.</li>
  </ul>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">C. Intuition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Think of data as cities connected by roads. UMAP builds a road network where each city connects to its nearest neighbours. Then it lays out the cities on a flat map, trying to keep the road network structure intact — cities connected by short roads should be close on the map.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Unlike t-SNE, UMAP preserves <em>some</em> of the global structure (relative positions of clusters), is <strong>faster</strong> (especially for large n), and has a well-defined <strong>transform</strong> method for new data.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">D. When to Use UMAP</h2>
  <table style="width:100%; border-collapse: collapse; margin-top:10px;">
    <tr style="background-color: #4B2E83; color: white;">
      <th style="padding:10px; text-align:left;">Aspect</th>
      <th style="padding:10px; text-align:left;">Details</th>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Best use cases</strong></td>
      <td style="padding:10px;">Large-scale visualisation; pre-processing for clustering; preserving global layout; new-data projection</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Assumptions</strong></td>
      <td style="padding:10px;">Data lie on or near a low-dimensional manifold; local distances are meaningful</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Strengths</strong></td>
      <td style="padding:10px;">Faster than t-SNE; preserves more global structure; has transform(); scales to millions</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Limitations</strong></td>
      <td style="padding:10px;">Still a nonlinear projection — distances are approximate; parameters matter; less established theory than PCA</td>
    </tr>
    <tr>
      <td style="padding:10px;"><strong>Common mistakes</strong></td>
      <td style="padding:10px;">Using default parameters without tuning; over-interpreting inter-cluster distances; not scaling data first</td>
    </tr>
  </table>
</div>

````python
# Example 7.1 — UMAP parameter tuning on Caravan
caravan = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Caravan.csv')
X_car = StandardScaler().fit_transform(caravan.select_dtypes(include=[np.number]))
purchase = (caravan['Purchase'] == 'Yes').values

params = [(5, 0.0), (15, 0.1), (50, 0.1), (50, 0.5)]
fig, axes = plt.subplots(1, 4, figsize=(22, 5))

for ax, (nn, md_val) in zip(axes, params):
    reducer = umap.UMAP(n_neighbors=nn, min_dist=md_val, random_state=42)
    Y = reducer.fit_transform(X_car)
    
    for val, color, label in [(False, UW_PURPLE, 'No'), (True, UW_GOLD, 'Yes')]:
        mask = purchase == val
        ax.scatter(Y[mask, 0], Y[mask, 1], c=color, label=label,
                   alpha=0.3 if not val else 0.8, s=5 if not val else 20)
    ax.set_title(f'nn={nn}, md={md_val}', fontsize=12, fontweight='bold', color=UW_DARK)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.legend(fontsize=8, title='Purchase')

fig.suptitle('UMAP: Effect of n_neighbors and min_dist on Caravan Data',
             fontsize=16, fontweight='bold', color=UW_DARK, y=1.02)
plt.tight_layout()
plt.show()

print("\nInterpretation: Small n_neighbors (5) emphasises very local structure,")
print("creating many small fragments. Larger n_neighbors (50) captures broader")
print("patterns. Small min_dist (0.0) packs clusters tightly; larger min_dist")
print("(0.5) spreads them out for easier visual inspection.")
````

  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Wage.csv</a></p>
  <p style="color:#555;"><strong>Question:</strong> Which method better reveals education-level groups in wage data?</p>

````python
# Example 7.2 — t-SNE vs UMAP on Wage
import time
wage = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Wage.csv')
X_w = StandardScaler().fit_transform(wage.select_dtypes(include=[np.number]))

# Subsample for t-SNE speed
np.random.seed(42)
idx = np.random.choice(len(X_w), 1000, replace=False)
X_w_sub = X_w[idx]
edu = wage['education'].values[idx]

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# t-SNE
t0 = time.time()
Y_tsne = TSNE(n_components=2, perplexity=30, random_state=42).fit_transform(X_w_sub)
t_tsne = time.time() - t0

# UMAP
t0 = time.time()
Y_umap = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42).fit_transform(X_w_sub)
t_umap = time.time() - t0

edu_colors = {'1. < HS Grad': '#C62828', '2. HS Grad': UW_GOLD,
               '3. Some College': '#1976D2', '4. College Grad': '#388E3C',
               '5. Advanced Degree': UW_PURPLE}

for ax, Y, title, t in [(axes[0], Y_tsne, 't-SNE', t_tsne),
                          (axes[1], Y_umap, 'UMAP', t_umap)]:
    for e, color in edu_colors.items():
        mask = edu == e
        ax.scatter(Y[mask, 0], Y[mask, 1], c=color, label=e.split('. ')[1],
                   alpha=0.5, s=20)
    ax.set_title(f'{title} ({t:.1f}s)', fontsize=14, fontweight='bold', color=UW_DARK)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.legend(fontsize=8, loc='best')

plt.tight_layout()
plt.show()

print(f"\nt-SNE: {t_tsne:.1f}s | UMAP: {t_umap:.1f}s")
print("UMAP is typically faster and often preserves more of the global arrangement.")
print("Both reveal that education levels partially cluster, though significant overlap")
print("exists — many other factors (age, experience) also drive wage variation.")
````

  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">OJ.csv</a></p>

````python
# Example 7.3 — UMAP + K-Means on OJ
oj = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/OJ.csv')
X_oj = StandardScaler().fit_transform(oj.select_dtypes(include=[np.number]))
purchase = oj['Purchase'].values

# UMAP embedding
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
X_umap = reducer.fit_transform(X_oj)

# K-Means in original space vs UMAP space
km_orig = KMeans(n_clusters=2, n_init=10, random_state=42).fit_predict(X_oj)
km_umap = KMeans(n_clusters=2, n_init=10, random_state=42).fit_predict(X_umap)

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# UMAP coloured by true labels
for brand, color in [('CH', UW_PURPLE), ('MM', UW_GOLD)]:
    mask = purchase == brand
    axes[0].scatter(X_umap[mask, 0], X_umap[mask, 1], c=color, label=brand,
                    alpha=0.4, s=15)
axes[0].set_title('UMAP (True Labels)', fontsize=13, fontweight='bold', color=UW_DARK)
axes[0].legend()

# K-Means in original space
for c, color in [(0, UW_PURPLE), (1, UW_GOLD)]:
    mask = km_orig == c
    axes[1].scatter(X_umap[mask, 0], X_umap[mask, 1], c=color,
                    label=f'Cluster {c}', alpha=0.4, s=15)
sil_orig = silhouette_score(X_oj, km_orig)
axes[1].set_title(f'K-Means (Original)\nSil = {sil_orig:.3f}',
                  fontsize=13, fontweight='bold', color=UW_DARK)
axes[1].legend()

# K-Means in UMAP space
for c, color in [(0, UW_PURPLE), (1, UW_GOLD)]:
    mask = km_umap == c
    axes[2].scatter(X_umap[mask, 0], X_umap[mask, 1], c=color,
                    label=f'Cluster {c}', alpha=0.4, s=15)
sil_umap = silhouette_score(X_umap, km_umap)
axes[2].set_title(f'K-Means (UMAP Space)\nSil = {sil_umap:.3f}',
                  fontsize=13, fontweight='bold', color=UW_DARK)
axes[2].legend()

for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()

print("\nInterpretation: Clustering in UMAP space can sometimes improve cluster quality")
print("because UMAP concentrates the meaningful structure into fewer dimensions.")
print("However, UMAP is a nonlinear transformation — interpret cluster boundaries")
print("with caution.")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">A. Learning Objectives</h2>

    <li>Explain <strong>reconstruction loss</strong> and how the network learns a compressed representation.</li>
    <li>Know the difference between autoencoders and PCA (nonlinear vs linear).</li>
    <li>Apply autoencoders for <strong>dimensionality reduction</strong>, <strong>anomaly detection</strong>, and <strong>latent space visualisation</strong>.</li>

</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">B. Definition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    An autoencoder is a neural network trained to <strong>reconstruct its own input</strong>. It consists of:
  </p>
  <ol style="color:#333; line-height:1.8;">
    <li><strong>Encoder:</strong> Maps input <strong>x</strong> ∈ ℝ<sup>p</sup> to a latent representation <strong>z</strong> ∈ ℝ<sup>d</sup> (where d ≪ p):  z = f<sub>enc</sub>(x)</li>
    <li><strong>Bottleneck:</strong> The layer with d neurons — the compressed representation.</li>
    <li><strong>Decoder:</strong> Maps <strong>z</strong> back to a reconstruction <strong>x̂</strong> ∈ ℝ<sup>p</sup>:  x̂ = f<sub>dec</sub>(z)</li>
  </ol>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    The <strong>reconstruction loss</strong> is typically MSE: L = (1/n) Σ ||x<sub>i</sub> − x̂<sub>i</sub>||².  
    When f<sub>enc</sub> and f<sub>dec</sub> are <em>linear</em> and the bottleneck has d neurons, the autoencoder learns the same subspace as PCA with d components. With nonlinear activations (ReLU, sigmoid), it captures <strong>nonlinear manifolds</strong>.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    A <strong>denoising autoencoder</strong> adds noise to the input and trains the network to recover the clean version — this prevents trivial identity mappings and learns more robust features.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">C. Intuition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    Imagine summarising a book in a single paragraph (encoder → bottleneck), then trying to reconstruct the full book from just that paragraph (decoder). The paragraph must capture the <em>essence</em> of the book. A good autoencoder learns what information is essential and what can be discarded.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>For anomaly detection:</strong> Train the autoencoder on "normal" data. When it encounters an anomaly, it cannot reconstruct it well — the reconstruction error spikes, flagging the anomaly.
  </p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">D. When to Use Autoencoders</h2>
  <table style="width:100%; border-collapse: collapse; margin-top:10px;">
    <tr style="background-color: #4B2E83; color: white;">
      <th style="padding:10px; text-align:left;">Aspect</th>
      <th style="padding:10px; text-align:left;">Details</th>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Best use cases</strong></td>
      <td style="padding:10px;">Nonlinear dimensionality reduction; anomaly detection; feature learning; denoising</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Assumptions</strong></td>
      <td style="padding:10px;">Data have redundancy that can be compressed; enough samples to train the network</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Strengths</strong></td>
      <td style="padding:10px;">Captures nonlinear structure; flexible architecture; can encode/decode new data</td>
    </tr>
    <tr style="border-bottom:1px solid #ddd;">
      <td style="padding:10px;"><strong>Limitations</strong></td>
      <td style="padding:10px;">Requires tuning (architecture, learning rate); less interpretable than PCA; needs more data; overfitting risk</td>
    </tr>
    <tr>
      <td style="padding:10px;"><strong>Common mistakes</strong></td>
      <td style="padding:10px;">Too-large bottleneck (identity mapping); not scaling input; training too long (overfitting); comparing directly to PCA without matching latent dimensions</td>
    </tr>
  </table>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 8.1 — Shallow Autoencoder: Credit Dataset</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Credit.csv</a></p>
  <p style="color:#555;"><strong>Question:</strong> Can a simple autoencoder compress 6 numeric features into 2 latent dimensions and reveal customer segments?</p>
</div>

````python
# Example 8.1 — Shallow autoencoder on Credit
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

credit = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Credit.csv')
X_ae = credit[['Income', 'Limit', 'Rating', 'Cards', 'Age', 'Balance']].values
X_ae = StandardScaler().fit_transform(X_ae).astype('float32')

# Build autoencoder
input_dim = X_ae.shape[1]
latent_dim = 2

encoder_input = keras.Input(shape=(input_dim,))
encoded = layers.Dense(16, activation='relu')(encoder_input)
encoded = layers.Dense(8, activation='relu')(encoded)
latent = layers.Dense(latent_dim, activation='linear', name='latent')(encoded)

decoded = layers.Dense(8, activation='relu')(latent)
decoded = layers.Dense(16, activation='relu')(decoded)
decoded = layers.Dense(input_dim, activation='linear')(decoded)

autoencoder = keras.Model(encoder_input, decoded)
encoder = keras.Model(encoder_input, latent)

autoencoder.compile(optimizer='adam', loss='mse')
history = autoencoder.fit(X_ae, X_ae, epochs=100, batch_size=32,
                          validation_split=0.2, verbose=0)

# Get latent representations
Z = encoder.predict(X_ae, verbose=0)

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Training loss
axes[0].plot(history.history['loss'], color=UW_PURPLE, linewidth=2, label='Train')
axes[0].plot(history.history['val_loss'], color=UW_GOLD, linewidth=2, label='Val')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('MSE Loss')
axes[0].set_title('Training History', fontsize=14, fontweight='bold', color=UW_DARK)
axes[0].legend()

# Latent space coloured by Balance
sc = axes[1].scatter(Z[:, 0], Z[:, 1], c=credit['Balance'], cmap='PuOr',
                     alpha=0.6, s=30, edgecolors='white')
axes[1].set_xlabel('Latent Dim 1')
axes[1].set_ylabel('Latent Dim 2')
axes[1].set_title('Autoencoder Latent Space', fontsize=14, fontweight='bold', color=UW_DARK)
plt.colorbar(sc, ax=axes[1], label='Balance')

# Compare to PCA
pca2 = PCA(n_components=2)
Z_pca = pca2.fit_transform(X_ae)
sc2 = axes[2].scatter(Z_pca[:, 0], Z_pca[:, 1], c=credit['Balance'], cmap='PuOr',
                      alpha=0.6, s=30, edgecolors='white')
axes[2].set_xlabel('PC1')
axes[2].set_ylabel('PC2')
axes[2].set_title('PCA (2 components)', fontsize=14, fontweight='bold', color=UW_DARK)
plt.colorbar(sc2, ax=axes[2], label='Balance')

plt.tight_layout()
plt.show()

print("\nInterpretation: The autoencoder latent space captures nonlinear patterns that")
print("PCA may miss. Both show the Balance gradient, but the autoencoder may reveal")
print("finer-grained structure due to its nonlinear capacity.")
````

````python
# Example 8.2 — Autoencoder anomaly detection on Default
default = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Default.csv')
X_def = default[['balance', 'income']].values
X_def = StandardScaler().fit_transform(X_def).astype('float32')
y_def = (default['default'] == 'Yes').values

# Train only on non-defaulters (normal data)
X_train_ae = X_def[~y_def]

# Build autoencoder
inp = keras.Input(shape=(2,))
enc = layers.Dense(8, activation='relu')(inp)
lat = layers.Dense(1, activation='linear')(enc)
dec = layers.Dense(8, activation='relu')(lat)
out = layers.Dense(2, activation='linear')(dec)

ae = keras.Model(inp, out)
ae.compile(optimizer='adam', loss='mse')
ae.fit(X_train_ae, X_train_ae, epochs=50, batch_size=64, verbose=0)

# Compute reconstruction error for ALL data
X_reconstructed = ae.predict(X_def, verbose=0)
recon_error = np.mean((X_def - X_reconstructed) ** 2, axis=1)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Distribution of reconstruction error by class
for val, color, label in [(False, '#388E3C', 'No Default'), (True, '#C62828', 'Default')]:
    mask = y_def == val
    axes[0].hist(recon_error[mask], bins=50, alpha=0.6, color=color, label=label, density=True)
axes[0].set_xlabel('Reconstruction Error (MSE)', fontsize=12)
axes[0].set_ylabel('Density', fontsize=12)
axes[0].set_title('Reconstruction Error Distribution', fontsize=14,
                  fontweight='bold', color=UW_DARK)
axes[0].legend()

# Scatter with colour = reconstruction error
threshold = np.percentile(recon_error, 95)
is_anomaly = recon_error > threshold
axes[1].scatter(default['balance'][~is_anomaly], default['income'][~is_anomaly],
                c='#388E3C', alpha=0.1, s=5, label='Normal')
axes[1].scatter(default['balance'][is_anomaly], default['income'][is_anomaly],
                c='#C62828', alpha=0.6, s=20, label='Anomaly (top 5%)')
axes[1].set_xlabel('Balance', fontsize=12)
axes[1].set_ylabel('Income', fontsize=12)
axes[1].set_title('Anomalies Detected by Autoencoder', fontsize=14,
                  fontweight='bold', color=UW_DARK)
axes[1].legend(fontsize=10)

plt.tight_layout()
plt.show()

# Check: what fraction of anomalies are actual defaulters?
anom_default_rate = y_def[is_anomaly].mean()
normal_default_rate = y_def[~is_anomaly].mean()
print(f"\nDefault rate among anomalies: {anom_default_rate:.1%}")
print(f"Default rate among normal points: {normal_default_rate:.1%}")
print(f"\nThe autoencoder successfully identifies high-risk customers — anomalies")
print(f"have a {anom_default_rate/normal_default_rate:.1f}x higher default rate.")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">B. Definition</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    An <strong>embedding</strong> is a learned mapping from a high-dimensional or discrete input space into a <strong>continuous, lower-dimensional vector space</strong> where <em>similar inputs are mapped to nearby points</em>.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Formally:</p>
  <p style="text-align:center; font-size:1.15em; color:#333;">
    f: X → ℝ<sup>d</sup>
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    where X can be words, images, products, users, or any entity, and d is the embedding dimension (typically 50–1000).
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">Key properties of a good embedding:</p>
  <ul style="color:#333; line-height:1.8;">
    <li><strong>Proximity = Similarity:</strong> Items that are semantically similar have vectors that are close (small cosine distance or Euclidean distance).</li>
    <li><strong>Arithmetic relationships:</strong> Famous example: <code>king − man + woman ≈ queen</code>. Directions in the embedding space can encode meaningful concepts like gender, tense, or size.</li>
    <li><strong>Dense and continuous:</strong> Unlike one-hot encoding (sparse, high-d), embeddings pack information into dense, low-d vectors.</li>
  </ul>

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">C. Intuition — Why Embeddings Matter</h2>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>The Address Book Analogy:</strong> Imagine you have 50,000 words in the English language. A one-hot encoding gives each word a unique ID — like giving each person a random phone number. You can't tell from two phone numbers whether the people know each other.
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    An embedding is like <strong>plotting everyone on a map based on where they live</strong>. Now neighbours on the map are neighbours in real life. "Cat" and "kitten" live on the same block. "Cat" and "refrigerator" live in different cities. And you can discover that the direction from "Paris" to "France" is the same as the direction from "Tokyo" to "Japan" — the map encodes the concept of "capital of".
  </p>
  <p style="font-size: 1.05em; color: #555; line-height: 1.7;">
    <strong>Where you encounter embeddings today:</strong>
  </p>
  <ul style="color:#333; line-height:1.8;">
    <li><strong>Search engines:</strong> Your query and web pages are both embedded — the engine returns pages whose vectors are closest to your query vector.</li>
    <li><strong>Recommendation systems:</strong> Netflix embeds users and movies into the same space — your next recommendation is the nearest unwatched movie to your user vector.</li>
    <li><strong>ChatGPT/LLMs:</strong> Every token is first converted to an embedding before the model processes it.</li>
    <li><strong>Image search:</strong> Photos are embedded so that similar images (same scene, same object) cluster together.</li>
  </ul>
</div>

  <h3 style="color: #4B2E83;">Example 9a.1 — Word Embeddings: Visualising Word Similarity in 2D</h3>
  <p style="color:#555;"><strong>Question:</strong> Can we see that semantically similar words are placed near each other in an embedding space?</p>
</div>

````python
# Example 9a.1 — Word embeddings visualised in 2D
# We'll create simple word embeddings using co-occurrence patterns
# and visualise them to see semantic clustering

# Define a small vocabulary with semantic groups
word_groups = {
    'Animals':   ['cat', 'dog', 'fish', 'bird', 'horse', 'mouse'],
    'Food':      ['pizza', 'burger', 'pasta', 'salad', 'rice', 'bread'],
    'Countries': ['france', 'japan', 'brazil', 'canada', 'italy', 'germany'],
    'Sports':    ['soccer', 'tennis', 'basketball', 'swimming', 'running', 'golf'],
    'Colors':    ['red', 'blue', 'green', 'yellow', 'purple', 'orange']
}

# Create synthetic embeddings that mimic real word2vec structure
# In real applications, you'd use pre-trained embeddings (GloVe, Word2Vec)
np.random.seed(42)
words = []
embeddings = []
labels = []

group_centres = {
    'Animals':   np.array([2.0, 3.0]),
    'Food':      np.array([-2.0, 2.5]),
    'Countries': np.array([0.0, -2.5]),
    'Sports':    np.array([3.0, -1.0]),
    'Colors':    np.array([-3.0, -1.0])
}

for group, group_words in word_groups.items():
    centre = group_centres[group]
    for w in group_words:
        words.append(w)
        labels.append(group)
        # Words in the same group cluster near the group centre
        embeddings.append(centre + np.random.normal(0, 0.5, 2))

embeddings = np.array(embeddings)

fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Plot 1: 2D embedding space with word labels
group_colors = {
    'Animals': UW_PURPLE, 'Food': UW_GOLD, 'Countries': '#C62828',
    'Sports': '#388E3C', 'Colors': '#1976D2'
}
for group in word_groups:
    mask = [l == group for l in labels]
    idx = [i for i, m in enumerate(mask) if m]
    axes[0].scatter(embeddings[idx, 0], embeddings[idx, 1],
                    c=group_colors[group], s=100, label=group,
                    alpha=0.8, edgecolors='white', zorder=5)
    for i in idx:
        axes[0].annotate(words[i], (embeddings[i, 0], embeddings[i, 1]),
                         fontsize=9, fontweight='bold',
                         xytext=(5, 5), textcoords='offset points')

axes[0].set_title('Word Embeddings in 2D Space',
                  fontsize=15, fontweight='bold', color=UW_DARK)
axes[0].set_xlabel('Embedding Dimension 1', fontsize=12)
axes[0].set_ylabel('Embedding Dimension 2', fontsize=12)
axes[0].legend(fontsize=10, title='Category')

# Plot 2: Similarity heatmap (cosine similarity)
from sklearn.metrics.pairwise import cosine_similarity
sim_matrix = cosine_similarity(embeddings)

# Select a subset for readability
subset_idx = [0, 1, 6, 7, 12, 13, 18, 19, 24, 25]  # 2 words per group
subset_words = [words[i] for i in subset_idx]
subset_sim = sim_matrix[np.ix_(subset_idx, subset_idx)]

sns.heatmap(subset_sim, annot=True, fmt='.2f', cmap='PuOr', center=0,
            xticklabels=subset_words, yticklabels=subset_words,
            ax=axes[1], vmin=-1, vmax=1)
axes[1].set_title('Cosine Similarity Between Word Embeddings',
                  fontsize=15, fontweight='bold', color=UW_DARK)

plt.tight_layout()
plt.show()

print("\nKey Observations:")
print("• Words in the same category (Animals, Food, etc.) cluster together")
print("• The cosine similarity is high (bright purple) within categories")
print("• The cosine similarity is low or negative between unrelated categories")
print("• This is the fundamental insight: embeddings encode MEANING as POSITION")
print("\nIn real NLP applications, these patterns emerge automatically from")
print("reading millions of sentences — words that appear in similar contexts")
print("end up with similar vectors.")
````

  <h3 style="color: #4B2E83;">Example 9a.2 — Embedding Arithmetic &amp; Analogies</h3>
  <p style="color:#555;"><strong>Question:</strong> Can we perform vector arithmetic on embeddings to discover relationships?</p>
</div>

````python
# Example 9a.2 — Embedding arithmetic (conceptual demonstration)
# Simulating the famous "king - man + woman = queen" analogy

# Create embeddings with built-in relational structure
np.random.seed(42)

# Gender offset vector
gender_offset = np.array([1.5, 0.3])
# Royalty offset vector
royalty_offset = np.array([0.2, 1.8])

concept_embeddings = {
    'man':    np.array([0.0, 0.0]),
    'woman':  np.array([0.0, 0.0]) + gender_offset,
    'king':   np.array([0.0, 0.0]) + royalty_offset,
    'queen':  np.array([0.0, 0.0]) + gender_offset + royalty_offset,
    'boy':    np.array([-0.8, -0.5]),
    'girl':   np.array([-0.8, -0.5]) + gender_offset,
    'prince': np.array([-0.8, -0.5]) + royalty_offset,
    'princess': np.array([-0.8, -0.5]) + gender_offset + royalty_offset,
}

# Add small noise for realism
for k in concept_embeddings:
    concept_embeddings[k] += np.random.normal(0, 0.1, 2)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Plot embeddings
for word, vec in concept_embeddings.items():
    axes[0].scatter(vec[0], vec[1], s=150, zorder=5,
                    c=UW_PURPLE if word in ['man','boy','king','prince'] else UW_GOLD,
                    edgecolors='white')
    axes[0].annotate(word, (vec[0], vec[1]), fontsize=12, fontweight='bold',
                     xytext=(8, 8), textcoords='offset points')

# Draw relationship arrows
for start, end, label in [('man', 'woman', 'gender'),
                            ('king', 'queen', 'gender'),
                            ('man', 'king', 'royalty'),
                            ('woman', 'queen', 'royalty')]:
    s, e = concept_embeddings[start], concept_embeddings[end]
    axes[0].annotate('', xy=e, xytext=s,
                     arrowprops=dict(arrowstyle='->', color='#C62828', lw=2, alpha=0.5))

axes[0].set_title('Embedding Space: Gender & Royalty Directions',
                  fontsize=14, fontweight='bold', color=UW_DARK)
axes[0].set_xlabel('Dimension 1')
axes[0].set_ylabel('Dimension 2')
axes[0].legend(['Male', 'Female'], fontsize=11)

# Demonstrate arithmetic
king = concept_embeddings['king']
man = concept_embeddings['man']
woman = concept_embeddings['woman']
queen_actual = concept_embeddings['queen']
queen_predicted = king - man + woman

all_words = list(concept_embeddings.keys())
all_vecs = np.array(list(concept_embeddings.values()))
dists = np.linalg.norm(all_vecs - queen_predicted, axis=1)
nearest = all_words[np.argmin(dists)]

axes[1].bar(all_words, dists, color=[UW_PURPLE if w != nearest else '#C62828' for w in all_words],
            edgecolor='white')
axes[1].set_ylabel('Distance from (king − man + woman)', fontsize=12)
axes[1].set_title(f'Nearest word to "king − man + woman" = "{nearest}"',
                  fontsize=14, fontweight='bold', color=UW_DARK)
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print(f"\nking - man + woman = predicted vector")
print(f"Nearest word to predicted vector: '{nearest}'")
print(f"Distance to actual 'queen': {np.linalg.norm(queen_actual - queen_predicted):.3f}")
print("\nThis demonstrates that embedding directions encode semantic relationships.")
print("The 'gender direction' (woman - man) can be added to any word to find")
print("its gendered counterpart. This works because the embedding space has")
print("learned these regularities from data.")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 9a.3 — Product Embeddings from Purchase Data: OJ Dataset</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">OJ.csv</a></p>
  <p style="color:#555;"><strong>Question:</strong> Can we create purchase-behaviour embeddings and visualise customer similarity?</p>
</div>

````python
# Example 9a.3 — Customer embeddings from purchase behaviour
oj = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/OJ.csv')
features_oj = ['PriceCH', 'PriceMM', 'DiscCH', 'DiscMM', 'LoyalCH',
               'SalePriceMM', 'SalePriceCH', 'PriceDiff', 'PctDiscMM', 'PctDiscCH']
X_oj = StandardScaler().fit_transform(oj[features_oj].values).astype('float32')

# Use an autoencoder to create 3D "purchase embeddings"
from tensorflow import keras
from tensorflow.keras import layers

inp = keras.Input(shape=(len(features_oj),))
e = layers.Dense(16, activation='relu')(inp)
z = layers.Dense(3, activation='linear', name='embedding')(e)  # 3D embedding
d = layers.Dense(16, activation='relu')(z)
o = layers.Dense(len(features_oj), activation='linear')(d)

ae = keras.Model(inp, o)
encoder = keras.Model(inp, z)
ae.compile(optimizer='adam', loss='mse')
ae.fit(X_oj, X_oj, epochs=80, batch_size=32, verbose=0)

# Get 3D embeddings, then project to 2D with UMAP
emb_3d = encoder.predict(X_oj, verbose=0)
emb_2d = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42).fit_transform(emb_3d)

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# Colour by brand purchased
for brand, color, label in [('CH', UW_PURPLE, 'Citrus Hill'), ('MM', UW_GOLD, 'Minute Maid')]:
    mask = oj['Purchase'] == brand
    axes[0].scatter(emb_2d[mask, 0], emb_2d[mask, 1], c=color, label=label,
                    alpha=0.4, s=15)
axes[0].set_title('Customer Embeddings (by Brand)',
                  fontsize=14, fontweight='bold', color=UW_DARK)
axes[0].legend(fontsize=11)
axes[0].set_xticks([]); axes[0].set_yticks([])

# Colour by loyalty
sc = axes[1].scatter(emb_2d[:, 0], emb_2d[:, 1], c=oj['LoyalCH'],
                     cmap='PuOr', alpha=0.5, s=15)
axes[1].set_title('Customer Embeddings (by CH Loyalty)',
                  fontsize=14, fontweight='bold', color=UW_DARK)
plt.colorbar(sc, ax=axes[1], label='Loyalty to Citrus Hill')
axes[1].set_xticks([]); axes[1].set_yticks([])

plt.tight_layout()
plt.show()

print("\nInterpretation: The autoencoder learned 3D 'purchase embeddings' that")
print("capture the essence of each customer's buying behaviour. When projected")
print("to 2D with UMAP, we see that:")
print("• Brand-loyal customers (high LoyalCH) cluster on one side")
print("• Brand-switchers occupy the middle ground")
print("• The embedding captures purchase patterns that go beyond any single feature")
print("\nThis is how modern recommendation systems work — each user gets an")
print("embedding vector, and similar users get similar recommendations.")
````

  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">D. Emerging Techniques (Brief Overview)</h2>

  <h3 style="color: #32006E;">Spectral Clustering</h3>
  <p style="color:#555; line-height:1.7;">Uses the eigenvalues of the similarity (affinity) matrix to perform dimensionality reduction before clustering. Particularly effective for non-convex cluster shapes.</p>

  <h3 style="color: #32006E;">Self-Organising Maps (SOMs)</h3>
  <p style="color:#555; line-height:1.7;">A neural network that produces a discretised, topology-preserving representation of the input space. Think of it as an "adaptive grid" that moulds to the data shape.</p>

  <h3 style="color: #32006E;">Variational Autoencoders (VAEs)</h3>
  <p style="color:#555; line-height:1.7;">Extend autoencoders by learning a <em>probability distribution</em> over the latent space (not just a point). This enables generating new, realistic data samples by sampling from the learned distribution.</p>

  <h3 style="color: #32006E;">Contrastive Learning Embeddings</h3>
  <p style="color:#555; line-height:1.7;">Self-supervised methods (e.g., SimCLR, BYOL) learn representations by pulling augmented views of the same data point together and pushing different data points apart. Used extensively in computer vision and NLP for pre-training.</p>
</div>

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 9.1 — Six Methods on College Data</h3>

  <p style="color:#555;"><strong>Question:</strong> How do PCA, MDS, t-SNE, UMAP, and an autoencoder compare when applied to the same dataset?</p>
</div>

````python
# Example 9.1 — Multi-method comparison on College
college = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/College.csv')
X_comp = StandardScaler().fit_transform(
    college.select_dtypes(include=[np.number])).astype('float32')
is_private = (college['Private'] == 'Yes').values

fig, axes = plt.subplots(2, 3, figsize=(20, 13))
axes = axes.ravel()
titles = ['PCA', 'Metric MDS', 't-SNE (perp=30)', 'UMAP', 'Autoencoder (2D)', 'K-Means on PCA']

# 1. PCA
Y_pca = PCA(n_components=2).fit_transform(X_comp)

# 2. MDS
D_comp = squareform(pdist(X_comp))
Y_mds = MDS(n_components=2, dissimilarity='precomputed', random_state=42,
            normalized_stress='auto').fit_transform(D_comp)

# 3. t-SNE
Y_tsne = TSNE(n_components=2, perplexity=30, random_state=42).fit_transform(X_comp)

# 4. UMAP
Y_umap = umap.UMAP(n_neighbors=15, min_dist=0.1, random_state=42).fit_transform(X_comp)

# 5. Autoencoder
inp = keras.Input(shape=(X_comp.shape[1],))
e = layers.Dense(32, activation='relu')(inp)
e = layers.Dense(16, activation='relu')(e)
z = layers.Dense(2, activation='linear')(e)
d = layers.Dense(16, activation='relu')(z)
d = layers.Dense(32, activation='relu')(d)
o = layers.Dense(X_comp.shape[1], activation='linear')(d)
ae_comp = keras.Model(inp, o)
enc_comp = keras.Model(inp, z)
ae_comp.compile(optimizer='adam', loss='mse')
ae_comp.fit(X_comp, X_comp, epochs=100, batch_size=32, verbose=0)
Y_ae = enc_comp.predict(X_comp, verbose=0)

# 6. K-Means on PCA
km_labels = KMeans(n_clusters=3, n_init=10, random_state=42).fit_predict(Y_pca)

embeddings = [Y_pca, Y_mds, Y_tsne, Y_umap, Y_ae, Y_pca]

for i, (ax, Y, title) in enumerate(zip(axes, embeddings, titles)):
    if i < 5:
        for val, color, label in [(True, UW_PURPLE, 'Private'), (False, UW_GOLD, 'Public')]:
            mask = is_private == val
            ax.scatter(Y[mask, 0], Y[mask, 1], c=color, label=label, alpha=0.5, s=15)
    else:
        cluster_colors = [UW_PURPLE, UW_GOLD, '#C62828']
        for c in range(3):
            mask = km_labels == c
            ax.scatter(Y[mask, 0], Y[mask, 1], c=cluster_colors[c],
                       label=f'Cluster {c}', alpha=0.5, s=15)
    ax.set_title(title, fontsize=13, fontweight='bold', color=UW_DARK)
    ax.legend(fontsize=8)
    ax.set_xticks([])
    ax.set_yticks([])

fig.suptitle('Method Comparison — College Dataset',
             fontsize=18, fontweight='bold', color=UW_DARK, y=1.01)
plt.tight_layout()
plt.show()

print("\nKey Observations:")
print("- PCA and MDS give similar linear projections")
print("- t-SNE and UMAP reveal tighter nonlinear clusters")
print("- The autoencoder finds a nonlinear 2D embedding similar to UMAP")
print("- K-Means on PCA scores gives interpretable clusters")
print("\nNo single method is best — choose based on your goal!")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 9.2 — Cluster Quality Across Methods: Credit Dataset</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Credit.csv</a></p>
  <p style="color:#555;"><strong>Question:</strong> Which embedding produces the best K-Means clusters (by silhouette score)?</p>
</div>

````python
# Example 9.2 — Silhouette comparison across embeddings
credit = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Credit.csv')
X_cr = StandardScaler().fit_transform(
    credit.select_dtypes(include=[np.number])).astype('float32')

methods = {}

# PCA
methods['PCA'] = PCA(n_components=2).fit_transform(X_cr)

# UMAP
methods['UMAP'] = umap.UMAP(n_neighbors=15, min_dist=0.1,
                              random_state=42).fit_transform(X_cr)

# t-SNE
methods['t-SNE'] = TSNE(n_components=2, perplexity=30,
                         random_state=42).fit_transform(X_cr)

# Autoencoder
inp = keras.Input(shape=(X_cr.shape[1],))
e = layers.Dense(16, activation='relu')(inp)
z = layers.Dense(2, activation='linear')(e)
d = layers.Dense(16, activation='relu')(z)
o = layers.Dense(X_cr.shape[1], activation='linear')(d)
ae = keras.Model(inp, o)
enc = keras.Model(inp, z)
ae.compile(optimizer='adam', loss='mse')
ae.fit(X_cr, X_cr, epochs=80, batch_size=32, verbose=0)
methods['Autoencoder'] = enc.predict(X_cr, verbose=0)

# Original space
methods['Original'] = X_cr

results = {}
for name, X_emb in methods.items():
    km = KMeans(n_clusters=3, n_init=10, random_state=42)
    labels = km.fit_predict(X_emb)
    sil = silhouette_score(X_emb, labels)
    results[name] = sil

fig, ax = plt.subplots(figsize=(10, 6))
colors = [UW_PURPLE, UW_GOLD, '#C62828', '#388E3C', '#1976D2']
bars = ax.bar(results.keys(), results.values(), color=colors, edgecolor='white', width=0.5)
for b in bars:
    ax.text(b.get_x() + b.get_width()/2, b.get_height() + 0.01,
            f'{b.get_height():.3f}', ha='center', fontsize=12, fontweight='bold')
ax.set_ylabel('Silhouette Score', fontsize=13)
ax.set_title('K-Means Silhouette Score Across Embeddings',
             fontsize=15, fontweight='bold', color=UW_DARK)
ax.set_ylim(0, max(results.values()) * 1.2)
plt.tight_layout()
plt.show()

print("\nInterpretation: Nonlinear methods (UMAP, autoencoder) often yield higher")
print("silhouette scores because they concentrate cluster structure into fewer")
print("dimensions. However, higher silhouette in a transformed space does not")
print("always mean the clusters are more meaningful — validate with domain knowledge.")
````

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">
  <h3 style="color: #4B2E83;">Example 9.3 — End-to-End Pipeline: Dimensionality Reduction → Clustering → Interpretation</h3>
  <p style="color:#555;"><strong>Dataset:</strong> <a href="https://github.com/PedroBSB/BDATA200/tree/main/Data">Hitters.csv</a></p>
  <p style="color:#555;"><strong>Question:</strong> Build a complete analysis pipeline: reduce dimensions, cluster, then profile the clusters.</p>
</div>

````python
# Example 9.3 — Full pipeline on Hitters
hitters = pd.read_csv('https://raw.githubusercontent.com/PedroBSB/BDATA200/main/Data/Hitters.csv').dropna()
X_hit = hitters.select_dtypes(include=[np.number])
feature_names = X_hit.columns.tolist()
X_hit_scaled = StandardScaler().fit_transform(X_hit).astype('float32')

# Step 1: UMAP to 2D
reducer = umap.UMAP(n_neighbors=20, min_dist=0.1, random_state=42)
X_2d = reducer.fit_transform(X_hit_scaled)

# Step 2: K-Means on UMAP embedding
km = KMeans(n_clusters=3, n_init=10, random_state=42)
labels = km.fit_predict(X_2d)

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# UMAP clusters
cluster_colors = [UW_PURPLE, UW_GOLD, '#C62828']
for c in range(3):
    mask = labels == c
    axes[0].scatter(X_2d[mask, 0], X_2d[mask, 1], c=cluster_colors[c],
                    label=f'Cluster {c}', alpha=0.6, s=40, edgecolors='white')
axes[0].set_title('UMAP + K-Means Clusters', fontsize=14, fontweight='bold', color=UW_DARK)
axes[0].legend()
axes[0].set_xticks([])
axes[0].set_yticks([])

# Step 3: Cluster profiles (radar/bar chart of means)
cluster_profiles = pd.DataFrame(X_hit_scaled, columns=feature_names)
cluster_profiles['Cluster'] = labels
means = cluster_profiles.groupby('Cluster').mean()

# Select top features by variance across clusters
feature_var = means.var(axis=0).sort_values(ascending=False)
top_features = feature_var.head(8).index.tolist()

x_pos = np.arange(len(top_features))
width = 0.25
for c in range(3):
    axes[1].bar(x_pos + c * width, means.loc[c, top_features],
                width=width, color=cluster_colors[c], label=f'Cluster {c}',
                edgecolor='white')
axes[1].set_xticks(x_pos + width)
axes[1].set_xticklabels(top_features, rotation=45, ha='right', fontsize=9)
axes[1].set_ylabel('Standardised Mean', fontsize=12)
axes[1].set_title('Cluster Profiles (Top 8 Features)', fontsize=14,
                  fontweight='bold', color=UW_DARK)
axes[1].legend(fontsize=9)
axes[1].axhline(0, color='grey', linewidth=0.5, linestyle='--')

# Salary distribution by cluster
for c in range(3):
    mask = labels == c
    salary = hitters.iloc[mask]['Salary']
    axes[2].hist(salary, bins=20, alpha=0.5, color=cluster_colors[c],
                 label=f'Cluster {c} (med=${salary.median():.0f}K)')
axes[2].set_xlabel('Salary ($1000s)', fontsize=12)
axes[2].set_ylabel('Count', fontsize=12)
axes[2].set_title('Salary Distribution by Cluster', fontsize=14,
                  fontweight='bold', color=UW_DARK)
axes[2].legend(fontsize=9)

plt.tight_layout()
plt.show()

print("\nPipeline Summary:")
print("1. UMAP reduced 16 features to 2D while preserving local structure.")
print("2. K-Means identified 3 player profiles in the UMAP space.")
print("3. Cluster profiles reveal distinct player types (e.g., veterans vs rookies).")
print("4. Salary distributions differ significantly across clusters,")
print("   validating that the clusters capture real differences.")
````
